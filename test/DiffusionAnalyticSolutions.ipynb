For 1D slab geometry with a uniform source, the solution to the diffusion equation is given as 
$$ \phi(x) = (1 - \frac{cosh(x/L)}{cosh(a/L)})\frac{Q}{\Sigma_a} $$
where $L = \sqrt{D/\Sigma_a}$ and $a$ is half the length of the domain. This uses the boundary condition $\phi(\pm a) = 0$

```python
def uniform_source(x, a, siga, D, Q):
    L = np.sqrt(D/siga)
    return (1 - np.cosh(x/L)/np.cosh(a/L))* Q/siga
```

```python
x = np.linspace(-5, 5, 1000)
a = 5
siga = 1
D = 1./(3*(1 + .5))
Q = 1
phi = uniform_source(x, a, siga, D, Q)
plt.plot(x, phi)
plt.show()
```
